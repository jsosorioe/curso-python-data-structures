```python
name = input("Enter file:")
if len(name) < 1:
    name = "/content/drive/MyDrive/b.txt"
handle = open(name)

counts = dict()
for line in handle:
  line = line.rstrip()
  if not line.startswith("From "):
    continue
  divword = line.split()
  hora = divword[5]
  hora = hora.split(":")
  divhora = hora[0]
  counts[divhora] = counts.get(divhora,0) + 1


lst= list()
for key, val in counts.items():
  newtupla = (val, key)
  print(key, val)
  lst.append(newtupla)

lst = sorted(lst, reverse = True)
print(lst)

for val, key in lst[:10]:
  print(key, val)
```

```python
import re
name = input("Enter file:")
if len(name) < 1:
    name = "/content/drive/MyDrive/muestra.txt"
       
handle = open(name)
s=0
sumar=0
lst =list()
for line in handle:
  line.rstrip()
  print(line)
  y = re.findall('([0-9]+)', line)
  if len(y) !=1:
    continue
  print(y)
  for num in y:
    sumar = sumar + int(num)
    print(sumar)
  lst.append(sumar)
print(s)
print(len(lst))
print(lst)
print(sumar)
w=0
for numm in lst:
  w=w+numm
print (w)
```
